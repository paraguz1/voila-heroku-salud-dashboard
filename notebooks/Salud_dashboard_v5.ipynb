```python
#Importamos datos desde CSV a pandas. Hay que subir el dataset antes a el repositorio GITHUB

df_trafico = pd.read_csv('https://raw.github.com/paraguz1/voila-heroku-salud-dashboard/master/Datos/BDD_datos.csv',
                            sep = ';', header = 0, encoding = 'latin1', decimal = ',')
#df_trafico.head()
```

```python
df_trafico_total = df_trafico[ (df_trafico.Canal == "Total") & 
                               (df_trafico.Dispositivo == "Total") &
                               (df_trafico.Edad == 'Total') &
                               (df_trafico.Género == 'Total') & (df_trafico.Mes == 'Total')]

fecha = df_trafico_total['Fecha'].values.tolist()
visitas = df_trafico_total['Visitas'].values.tolist()
visitas_recurrentes = df_trafico_total['Visitas recurrentes'].values.tolist()
pedidos = df_trafico_total['Pedidos'].values.tolist()
precio_medio = df_trafico_total['Precio medio'].values.tolist()
num_art_medio = df_trafico_total['Número medio artículos'].values.tolist()
sus_newsletter = df_trafico_total['Suscripciones newletter'].values.tolist()
sus_web =df_trafico_total['Suscripciones web'].values.tolist()
```

```python
def response_fig(change):
          
     
        with fig_resp.batch_update():
            
                df_filtered = df_solicitudes[(df_solicitudes['Unidad'] == str(filter_canal.value)) 
                                                                & (df_solicitudes['modalidad'] == str(filter_modalidad.value))
                                                                & (df_solicitudes['producto'] == str(filter_producto.value))
                                                                & (df_solicitudes['target'] == str(filter_target.value))
                                                                & (df_solicitudes['NP'] == 'Total' )
                                                                & (df_solicitudes['baja_vto'] == 'Total' )
                                                                & (df_solicitudes['baja_egas_permitida'] == 'Total' )
                                                                & (df_solicitudes['baja_excepcional'] == 'Total' )
                                                                & (df_solicitudes['fecha'] == str(filter_fecha.value))]
                
                
                
                
                #To update nodes labels and positions with the filtered data (some nodes can disappear when filtering)
                source = df_filtered['nivel_incio'].values.tolist()
                source_labels = df_filtered['des_nivel_inicio'].values.tolist()
                
                target = df_filtered['nivel_final'].values.tolist()
                target_labels = df_filtered['des_nivel_final'].values.tolist()
                
                source_lab = list (set (list (zip(source, source_labels))))
                target_lab = list (set (list (zip(target, target_labels))))
                
                #We obtain unique nodes and labels and order by node number
                nodes = sorted (list (set (source_lab + target_lab)), key = itemgetter(0))
                
                #Obtain new labels, node colors and x,y nodes positions (with filtered nodes, only the nodes that remain after filtering)
                labels_filtered = [elem[1] for elem in nodes]
                nodes_filtered = [ elem[0] for elem in nodes]
                
                asegurados_node_inicio_filtered = df_filtered[['asegurados', 'nivel_incio']].groupby(['nivel_incio']).sum()['asegurados'].values.tolist()
                asegurados_node_final_filtered = df_filtered[(df_filtered['nivel_final'] == 17) | (df_filtered['nivel_final'] == 18) | (df_filtered['nivel_final'] == 19) ][['asegurados', 'nivel_final']].groupby(['nivel_final']).sum()['asegurados'].values.tolist()

                asegurados_node_filtered = asegurados_node_inicio_filtered + asegurados_node_final_filtered
                
                
                node_colors_filtered = []
                node_x_position_filtered = []
                node_y_position_filtered = []
                
                
                for elem in nodes_filtered:
                    node_colors_filtered.append(node_colors[elem])
                    node_x_position_filtered.append(node_x_position[elem])
                    node_y_position_filtered.append(node_y_position[elem])
                    
                
                fig_resp.data[0].node['y'] = node_y_position_filtered
                fig_resp.data[0].node['x'] = node_x_position_filtered
                fig_resp.data[0].node['color'] = node_colors_filtered
                fig_resp.data[0].node['label'] = labels_filtered
                fig_resp.data[0].node['customdata'] = asegurados_node_filtered
                
                nodes_index = list(range(len(nodes_filtered)))
                nodes_dict = dict (zip (nodes_filtered, nodes_index))
                
                source_filtered = []
                target_filtered = []
                
                nivel_inicio_filtered = df_filtered['nivel_incio'].values.tolist()
                nivel_final_filtered = df_filtered['nivel_final'].values.tolist()
                
                #To re-index nodes, when nodes disappear after filtering then whe have to reorder and change node number in 
                #source and target
                for elem in nivel_inicio_filtered:
                    source_filtered.append(nodes_dict[elem])
                    
                for elem in nivel_final_filtered:
                    target_filtered.append(nodes_dict[elem])  
                    
                #Add new re-ordered nodes for source and target     
                fig_resp.data[0].link['source'] = source_filtered
                
                fig_resp.data[0].link['target'] = target_filtered
              
                fig_resp.data[0].link['value']  = df_filtered['solicitudes'].values.tolist()
                fig_resp.data[0].link['color'] = df_filtered['flow_color'].values.tolist()  
                
                fig_resp.data[0].link['customdata'] = df_filtered['asegurados'].values.tolist()
                
                
                
                
                
```
